```python
# Loading the first 100 rows of the dataset to explore and understand its structure
salary_df = pd.read_excel('./salary_data_states.xlsx')

# Displaying the first few rows of the dataset to understand its structure
salary_df.head()
```

```python
filtered_df = salary_df[['EMPLOYER_NAME', 'PREVAILING_WAGE_PER_YEAR', 'PAID_WAGE_PER_YEAR', 'JOB_TITLE', 'WORK_STATE']].dropna()
```

```python
filtered_df['PREVAILING_WAGE_PER_YEAR'] = pd.to_numeric(filtered_df['PREVAILING_WAGE_PER_YEAR'], errors='coerce')
filtered_df['PAID_WAGE_PER_YEAR'] = pd.to_numeric(filtered_df['PAID_WAGE_PER_YEAR'], errors='coerce')
```

```python
# Clustered Bar Chart: Average salaries across employers
avg_salary_employer = filtered_df.groupby('EMPLOYER_NAME')['PAID_WAGE_PER_YEAR'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=avg_salary_employer.values, y=avg_salary_employer.index, palette='viridis')
plt.xlabel('Average Paid Wage per Year')
plt.ylabel('Employer Name')
plt.title('Top 10 Employers with Highest Average Salaries')
plt.show()
```

```python
# Violin Plot: Distribution of salaries within work states
plt.figure(figsize=(12, 8))
sns.violinplot(data=filtered_df, x='WORK_STATE', y='PAID_WAGE_PER_YEAR', palette='muted')
plt.xlabel('Work State')
plt.ylabel('Paid Wage per Year')
plt.title('Distribution of Salaries by State')
plt.xticks(rotation=45)
plt.show()
```

```python
# Violin Plot: Distribution of salaries across job titles
plt.figure(figsize=(14, 8))
filtered_top_jobs = filtered_df[filtered_df['JOB_TITLE'].isin(filtered_df['JOB_TITLE'].value_counts().head(10).index)]
sns.violinplot(data=filtered_top_jobs, x='JOB_TITLE', y='PAID_WAGE_PER_YEAR', hue='JOB_TITLE', palette='pastel', legend=False)
plt.xlabel('Job Title')
plt.ylabel('Paid Wage per Year')
plt.title('Distribution of Salaries by Job Title (Top 10)')
plt.xticks(rotation=45)
plt.show()
```

```python
import networkx as nx

# Network Graph: Relationships between employer, work state, and salary levels
subset_df = filtered_df.head(20)
G = nx.Graph()

for _, row in subset_df.iterrows():
    employer = row['EMPLOYER_NAME']
    state = row['WORK_STATE']
    salary = row['PAID_WAGE_PER_YEAR']

    G.add_node(employer, label='Employer')
    G.add_node(state, label='Work State')
    G.add_node(salary, label='Salary')
    
    G.add_edge(employer, state)
    G.add_edge(state, salary)

plt.figure(figsize=(10, 10))
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=2000, font_size=10, font_weight='bold', edge_color='gray')
plt.title('Network Graph of Employers, Work States, and Salary Levels')
plt.show()
```

```python

data = {
    'COMPANY_SIZE': ['Startup', 'Large Corporation', 'Startup', 'Large Corporation', 'Startup', 'Large Corporation'],
    'PAID_WAGE_PER_YEAR': [55000, 75000, 52000, 78000, 50000, 80000]
}
salary_df = pd.DataFrame(data)

# Box Plot: Comparing salaries at Startups vs Large Corporations
plt.figure(figsize=(12, 8))
sns.boxplot(data=salary_df, x='COMPANY_SIZE', y='PAID_WAGE_PER_YEAR', palette='Set2')
plt.xlabel('Company Size')
plt.ylabel('Paid Wage per Year')
plt.title('Comparison of Salaries at Startups vs Large Corporations')
plt.tight_layout()
plt.show()
```
