```python
import pdfplumber
from IPython.display import display, Markdown

GroupProject_pdf = './DataSummary.pdf'

with pdfplumber.open(GroupProject_pdf) as pdf:
    pdf_text = ''
    
    for page in pdf.pages:
        pdf_text += page.extract_text(layout=True)

display(Markdown(f"### Extracted PDF Text:\n\n{pdf_text}"))

```
